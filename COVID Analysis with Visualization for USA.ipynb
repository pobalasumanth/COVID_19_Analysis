# COVID 19 Analysis 
## This analysis focus on the United States country

```python
import pandas as pd
```

```python
df=pd.read_csv('input/Covid_Confirmed.csv')
df1=pd.read_csv('input/Covid_Death.csv')
df2=pd.read_csv('input/Covid_recovered.csv')
```

```python
df.head()
```

```python
df1.head()
```

```python
df2.head()
```

```python
import matplotlib.pyplot as plt
```

```python
df=df.drop(["Province/State","Lat","Long"],axis=1)
```

```python
df.head()
```

```python
df=df.groupby(df["Country/Region"]).aggregate('sum')
```

```python
df2=df2.drop(["Province/State","Lat","Long"],axis=1)
df1=df1.drop(["Province/State","Lat","Long"],axis=1)
```

```python
df1=df1.groupby(df1["Country/Region"]).aggregate('sum')
df2=df2.groupby(df2["Country/Region"]).aggregate('sum')
```

```python
print(df2)
```

```python
df=df.T
df1=df1.T
df2=df2.T
```

```python
new_cases=df.copy()
for day in range(1, len(df)):
    new_cases.iloc[day]=df.iloc[day]-df.iloc[day-1]
```

```python
growth_rate=df.copy()
for day in range(1,len(df)):
    growth_rate.iloc[day]= (new_cases.iloc[day]/df.iloc[day-1]) * 100
growth_rate.head()
```

```python
active_cases=df.copy()
for day in range(1,len(df)):
    active_cases.iloc[day]= df.iloc[day]= df2.iloc[day]
active_cases.tail()
```

```python
overall_growth_rate =df.copy()

for day in range(1,len(df)):
    overall_growth_rate.iloc[day]=((active_cases.iloc[day] - active_cases.iloc[day-1]) /active_cases.iloc[day-1])*100
```

```python
overall_growth_rate['US'].tail(10)
```

```python
death_rate = df.copy()
for day in range(1,len(df)):
    death_rate.iloc[day]=(df1.iloc[day]/df.iloc[day])*100
death_rate
```

```python
death_rate["US"].tail(10)
```

```python
hospitalization_rate_estimate = 0.05
hostpitalization_needed=df.copy()
for day in range(1,len(df)):
    hostpitalization_needed.iloc[day]= active_cases.iloc[day] * hospitalization_rate_estimate

hostpitalization_needed["US"].tail()  
        
```
